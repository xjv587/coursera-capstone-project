```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]

target = concrete_data['Strength']
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1]
n_cols
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = baseline_model()

model.fit(X_train, y_train, epochs=50)
```

```python
y_pred = model.predict(X_test)

print('Mean squared error of the test data is:', mean_squared_error(y_test, y_pred))
```

```python
mse = []

for i in range(50):
    model = baseline_model()
    model.fit(X_train, y_train, epochs=50)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
mean = np.mean(mse)
std = np.std(mse)

print("Mean of MSE is:", mean, ".    Standard deviation of MSE is:", std)
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
X_train_norm = (X_train - X_train.mean()) / X_train.std()
X_test_norm = (X_test - X_test.mean()) / X_test.std()
y_train_norm = (y_train - y_train.mean()) / y_train.std()
y_test_norm = (y_test - y_test.mean()) / y_test.std()
```

```python
model.fit(X_train_norm, y_train_norm, epochs=50)
```

```python
y_pred_norm = model.predict(X_test_norm)

print('MSE of the normalized test data is:', mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mse_b = []

for i in range(50):
    model = baseline_model()
    model.fit(X_train_norm, y_train_norm, epochs=50)
    y_pred_norm = model.predict(X_test_norm)
    mse_b.append(mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mean2 = np.mean(mse_b)
std2 = np.std(mse_b)

print("Mean of MSE based on normalized predictors is:", mean2, ".    Standard deviation of MSE based on normalized predictors is:", std2)
print("Whereas in part A, Mean of MSE is:", mean, ". Standard deviation of MSE is:", std)
```

```python
model.fit(X_train_norm, y_train_norm, epochs=100)
```

```python
y_pred_norm = model.predict(X_test_norm)

print('Mean Squared error of test data with 100 epochs is:', mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mse_c = []

for i in range(100):
    model = baseline_model()
    model.fit(X_train_norm, y_train_norm, epochs=100)
    y_pred_norm = model.predict(X_test_norm)
    mse_c.append(mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mean3 = np.mean(mse_c)
std3 = np.std(mse_c)

print("With 100 epochs, mean of MES_norm is:", mean3, ".     Standard deviation is:", std3)
print("In part B, 50 epochs, mean of MSE_norm is:", mean2, ".    Standard deviation is:", std2)
print("In part A, mean of MSE is:", mean, ". Standard deviation is:", std)
```

```python
# define multilayer model
def ml_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model1 = ml_model()
model1.fit(X_train_norm, y_train_norm, epochs=50)
```

```python
y_pred_norm = model1.predict(X_test_norm)

print('MSE of multilayer neural network of norm_test_data with 50 epochs is:', mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mse_d = []

for i in range(50):
    model1.fit(X_train_norm, y_train_norm, epochs=50)
    y_pred_norm = model1.predict(X_test_norm)
    mse_d.append(mean_squared_error(y_test_norm, y_pred_norm))
```

```python
mean4 = np.mean(mse_d)
std4 = np.std(mse_d)

print("In part D, 3-hidden-layer 50 epochs mean of MES_norm is:", mean4, ". Standard deviation is:", std4)
print("In part C, 1-hidden-layer 100 epochs, mean of MES_norm is:", mean3, ". Standard deviation is:", std3)
print("In part B, 1-hidden-layer 50 epochs, mean of MSE_norm is:", mean2, ". Standard deviation is:", std2)
print("In part A, 1-hidden-layer 50 epochs, mean of MSE is:", mean, ". Standard deviation is:", std)
```
